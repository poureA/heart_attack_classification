```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
scaler = StandardScaler()
```

```python
label = df['DEATH_EVENT']
df = df.drop(['DEATH_EVENT'],axis=1)
scaled_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
scaled_df.head(10)
```

```python
xtr , xts , ytr , yts = train_test_split(df,label,test_size=0.2,random_state=32)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
alg = GradientBoostingClassifier()
alg.fit(xtr,ytr)
alg.score(xtr,ytr)*100
```

```python
alg.score(xts,yts)*100
```

```python
alg.score(df,label)*100
```

```python
cross = cross_val_score(alg,df,label,cv=15)
res = np.sqrt(cross)
np.mean(res)*100
```
