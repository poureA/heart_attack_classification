```python
import pandas as pd
import numpy as np
```

```python
file =pd.read_csv('F:datasets\\healthcare-dataset-stroke-data.csv')
df = file.copy()
df.sample(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
bmi = df['bmi'].to_numpy().reshape(-1,1)
label = df['stroke']
df = df.drop(['id','stroke','bmi'],axis=1)
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputed_bmi = pd.DataFrame(imputer.fit_transform(bmi),columns=['bmi'])
df = pd.concat([df,imputed_bmi],axis=1)
df.isnull().sum()
```

```python
df.info()
```

```python
for i in df.columns :
    print(i)
    print(df[i].value_counts(ascending=False))
```

```python
encoded_df = pd.get_dummies(df)
encoded_df.info()
```

```python
encoded_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(encoded_df),columns=encoded_df.columns)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr ,yts = train_test_split(scaled_df,label,test_size=0.2,shuffle=True)
print('spliting done')
```

```python
from sklearn.svm import LinearSVC
alg = LinearSVC().fit(xtr,ytr)
alg.score(xtr,ytr)*100
```

```python
alg.score(xts,yts)*100
```

```python
alg.score(scaled_df,label)*100
```
